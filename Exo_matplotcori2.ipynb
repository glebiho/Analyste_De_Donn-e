```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 10)
```

```python
y = x**2
```

```python
z = x**3
```

```python
# Courbe y=fct(x)
plt.plot(x, y);
```

```python
#2 modifications
plt.plot(x, y, c='red', lw=10); 
```

```python
#### 2 courbes
plt.plot(x, y, c='red', lw=5, label="Courbe de x en fonction de y");
plt.plot(x, z, c='black', lw=2.5, label="Courbe de x en fonction de z");
plt.legend();
plt.title("Courbes d'évolution"); 
plt.xlabel("Axe des abscisses (x)");
plt.ylabel("Axe des ordonnées (y)");
```

```python
#4 Boxplot de x
plt.boxplot(x);
```

```python
#5 Histogramme de x
plt.hist(x);
```

```python
#6 Nuage de point entre x et y
plt.scatter(x, y)
plt.xlabel("Axe des abscisses (x)");
plt.ylabel("Axe des ordonnées (y)");
plt.title("Nuage de point entre x et y", fontsize=20); 
```
