```python
#11
liste_qual = ['SEXE', 'TDT', 'ECG', 'ANGINE', 'PENTE', 'CÅ’UR', 'GAJ']
liste_qual
```

```python
liste_quant = ['AGE', 'CHOLESTEROL', 'PAR', 'FCMAX', 'DEPRESSION']
```

```python
for i in liste_quant:
    plt.hist(coeur[i], edgecolor='black', label=i)
    plt.title(i)
    plt.legend()
    plt.show()
    
```

```python
colors = ['blue', 'red', 'orange', 'cyan', 'grey']

for index, libelle in enumerate(liste_quant):
    plt.figure(figsize=(10, 5))
    plt.hist(coeur[libelle], color=colors[index], label=libelle, edgecolor='black')
    plt.legend()


plt.show()
```

```python
colors_1 = ['blue', 'red', 'orange', 'cyan', 'grey', 'green', 'brown']

for index, libelle in enumerate(liste_qual):
    plt.figure(figsize=(10, 5))
    x=coeur[libelle].value_counts()
    x.plot(kind='bar', label=libelle, color=colors_1[index])
#    plt.bar(coeur[libelle], color=colors[index], label=libelle, edgecolor='black', height=, width=5)
    plt.legend()


plt.show()
```

```python
colors_1 = ['blue', 'red', 'orange', 'cyan', 'grey', 'green', 'brown']
i=1
plt.figure(figsize=(10, 5))
for index, libelle in enumerate(liste_qual):
    
    plt.subplot(2, 3, i)
    x=coeur[libelle].value_counts()
    x.plot(kind='bar', label=libelle, color=colors_1[index])
#    plt.bar(coeur[libelle], color=colors[index], label=libelle, edgecolor='black', height=, width=5)
    plt.legend()
    i+=1
    plt.show()
```

```python
plt.figure(figsize=(30, 15))
plt.gcf().subplots_adjust(wspace=1, hspace=1)
i=1
for var in liste_qual:
    plt.subplot(3,3,i)
    x=coeur[var].value_counts()
    x.plot(kind='bar')
    plt.title(var, fontsize=40)
    #plt.xlabel(fontsize=20)
    plt.xticks(fontsize=20, rotation=45)
    #plt.ylabel(fontsize=20)
    plt.yticks(fontsize=20)

    i+=1
```
