d'après les informations precedente nous voyons que dans notre etude se porte sur 505 clients avec 11 critères different qui sont: Name, Phone Number, Email, Address, Country, Postal code,Last date of connection, Last time of connection, Age, Gender,Customer spendings et a part l'age et les depences consomateur le reste des donné sont de valeur qualitatives

```python
#Calculer la médiane et la moyenne des colonnes
print("mediane age",dataset["Age"].median())
print("mediane Customer spendings",dataset["Customer spendings"].median())

print("moyenne age",dataset["Age"].mean())
print("moyenne Customer spendings",dataset["Customer spendings"].mean())

```

```python
# mediane d'age pour chaque pays
dataset.groupby(['Country'])['Age'].median()
```

```python
#recoder l’age en variable qualitative ne faisant des intervalle de 20

def age1(age):
    if age < 18:
        return 'mineur'
    elif age < 38:
        return 'jeune'
    elif age < 58:
        return 'adulte'
    else:
        return 'vieux'

dataset['Age'] = dataset['Age'].map(age1)
dataset
```

```python
df2 = dataset.groupby('Country')['Customer spendings'].sum()
print(df2)
```

```python
countries = dataset['Country']
#plt.rcParams['figure.figsize'] = (30, 30)
#plt.rcParams['font.size'] = '17'
plt.xticks(rotation = '45')
plt.title("les dépenses des clients par pays")
plt.xlabel('Country')
plt.ylabel('Customer spendings')
#plt.bar(dataset['Country'], dataset['Customer spendings'])
dataset.groupby('Country')['Customer spendings'].sum().plot(kind='bar', x='Country',y='Customer spendings',figsize=(20,10))
#df2.plot(kind='bar')
```

```python
#Nettoyer les lignes avec moins de 10 € de dépenses par client : supprimez toutes les
#lignes d’utilisateurs ayant dépensé moins de 10 € sur le site,

dataset.drop( dataset[ (dataset['Customer spendings'] < 10)].index, inplace=True)
```

```python
#Nettoyer les doublons : supprimer toutes les lignes qui apparaissent plus d’une fois
#dans les données,
dataset.drop_duplicates( keep = 'first' , inplace = True , ignore_index = False )

```

```python
dataset.drop(columns= ['Name', 'Phone Number', 'Email', 'Address', 'Postal code',
                       'Last date of connection', 'Last time of connection'], axis=1, inplace=True)
```

```python
#Écrire les données nettoyées dans un nouveau fichier CSV avec uniquement les
#colonnes suivantes :
#○ « Country »
#○ « Age »
#○ « Gender »
#○ « Customer spendings »
#Le livrable est un fichier Jupyter Notebook contenant le travail effectué
data =pd.DataFrame(dataset)
data.to_csv("data.csv")
data = pd.read_csv("data.csv")
data
```
