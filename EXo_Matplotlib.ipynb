```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 10)
y = x**2
z = x**3
print(x)
```

```python
# 1. Construire une courbe avec x en abscisse et y en ordonnée
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
# 1.Changer l'épaisseur et la couleur de la ligne du graphique précédemment créé
avec le paramètre "c" et "lw"
plt.figure()
plt.plot(x, y, c='red',lw= 4)
plt.show()
```

```python
# 3. Construire sur le même graphique deux courbes :
plt.plot(x, y, label='quadratique')
plt.plot(x, z, label='cubique')
plt.title('Excercice 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
```
